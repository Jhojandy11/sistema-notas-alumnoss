```python
import json

# ================== SISTEMA DE NOTAS ==================

class Alumno:
    def __init__(self, aid, nombre, nota1, nota2, nota3):
        self.aid = aid
        self.nombre = nombre
        self.nota1 = nota1
        self.nota2 = nota2
        self.nota3 = nota3
        self.promedio = round((nota1 + nota2 + nota3) / 3, 2)

    def a_dict(self):
        return {
            "aid": self.aid,
            "nombre": self.nombre,
            "nota1": self.nota1,
            "nota2": self.nota2,
            "nota3": self.nota3,
            "promedio": self.promedio
        }

    @staticmethod
    def desde_dict(d):
        return Alumno(d["aid"], d["nombre"], d["nota1"], d["nota2"], d["nota3"])


# Lista de alumnos
alumnos = []
siguiente_id = 1
archivo = "alumnos.json"


# ================== OPCIÃ“N 1: REGISTRAR ==================

def registrar_alumno():
    global siguiente_id

    print("\n--- Registrar alumno ---")
    nombre = input("Nombre del alumno: ")

    try:
        n1 = float(input("Nota 1: "))
        n2 = float(input("Nota 2: "))
        n3 = float(input("Nota 3: "))
    except ValueError:
        print(" Error: Debes ingresar nÃºmeros.")
        return

    nuevo = Alumno(siguiente_id, nombre, n1, n2, n3)
    alumnos.append(nuevo)

    print(f"âœ” Alumno '{nuevo.nombre}' registrado con ID {nuevo.aid}. Promedio: {nuevo.promedio}")

    siguiente_id += 1


# ================== OPCIÃ“N 2: LISTAR ==================

def listar_alumnos():
    print("\n--- Lista de alumnos ---")
    if len(alumnos) == 0:
        print("No hay alumnos registrados.")
        return

    for a in alumnos:
        print(f"ID {a.aid} | {a.nombre} | Notas: {a.nota1}, {a.nota2}, {a.nota3} | Promedio: {a.promedio}")


# ================== OPCIÃ“N 3: ELIMINAR ==================

def eliminar_alumno():
    if len(alumnos) == 0:
        print("No hay alumnos para eliminar.")
        return

    try:
        aid = int(input("Ingrese el ID del alumno a eliminar: "))
    except:
        print(" Debes ingresar un nÃºmero.")
        return

    for a in alumnos:
        if a.aid == aid:
            alumnos.remove(a)
            print(f"ðŸ—‘ Alumno con ID {aid} eliminado.")
            return

    print(" No existe un alumno con ese ID.")


# ================== OPCIÃ“N 4: GUARDAR ==================

def guardar_datos():
    data = [a.a_dict() for a in alumnos]

    with open(archivo, "w") as f:
        json.dump({
            "siguiente_id": siguiente_id,
            "alumnos": data
        }, f, indent=4)

    print(" Datos guardados correctamente.")


# ================== MENÃš PRINCIPAL ==================

def mostrar_menu():
    print("""
======= MENÃš SISTEMA DE NOTAS =======
1. Registrar alumno
2. Listar alumnos
3. Eliminar alumno
4. Guardar datos
0. Salir
=====================================
""")

def main():
    while True:
        mostrar_menu()
        op = input("OpciÃ³n: ")

        if op == "1":
            registrar_alumno()

        elif op == "2":
            listar_alumnos()

        elif op == "3":
            eliminar_alumno()

        elif op == "4":
            guardar_datos()

        elif op == "0":
            print(" Saliendo del sistema...")
            break

        else:
            print(" OpciÃ³n invÃ¡lida.")


# Ejecutar
main()
```
